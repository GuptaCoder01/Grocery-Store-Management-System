**Importing The Dependencies**


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection & Data Processing**

```python
#Loading the dataset to a pandas DataFrame
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)
```

```python
sonar_data.head()
```

```python
#Number of Rows & Columns
sonar_data.shape
```

```python
#Describe --> Statistical Measures Of The Data.
sonar_data.describe()
```

```python
sonar_data.value_counts()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
sonar_data.groupby(60).median()
```

```python
# seprating data and labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training --> Logistics Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuaracy on training data
X_train_predtiction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predtiction, Y_train)
```

```python
print('Accuracy on training data :', training_data_accuracy)
```

```python
#Accuaracy on training data
X_test_predtiction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predtiction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
